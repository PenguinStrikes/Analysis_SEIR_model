## SEIR Analysis
This notebook loads in the SEIR modelling class and tweaks the beta, gamma and sigma parameters in order to understand the potential impact on the population model.

```python
%load_ext autoreload
%autoreload 2

import os
import sys

PROJ_ROOT = os.path.abspath(os.path.join(os.pardir))
sys.path.append(os.path.join(PROJ_ROOT, 'src'))

import seir
```

```python
o = seir.SEIR_modelling(N=100)
o.SEIR_solver()
o.plot_model(i=119)
```

```python
o.plot_animation(skip_frames=True)
```

```python
p = seir.SEIR_modelling(N=100, beta=0.5)
p.SEIR_solver()
p.plot_model(i=40)

p = seir.SEIR_modelling(N=100, beta=0.8)
p.SEIR_solver()
p.plot_model(i=40)

p = seir.SEIR_modelling(N=100, beta=2.0)
p.SEIR_solver()
p.plot_model(i=40)
```

```python
p = seir.SEIR_modelling(N=100, gamma=0.1)
p.SEIR_solver()
p.plot_model(i=119)

p = seir.SEIR_modelling(N=100, gamma=0.05)
p.SEIR_solver()
p.plot_model(i=119)

p = seir.SEIR_modelling(N=100, gamma=0.01)
p.SEIR_solver()
p.plot_model(i=119)
```
